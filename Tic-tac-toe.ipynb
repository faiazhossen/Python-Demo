```python
from IPython.display import clear_output

def display_board(board):
    print("  |   |  ")
    print(board[7]+" | "+ board[8] + " | "+ board[9])
    print("  |   |  ")
    print('_ '*5)
    
    print("  |   |  ")
    print(board[4]+" | "+ board[5] + " | "+ board[6])
    print("  |   |  ")
    print('_ '*5)
    
    print("  |   |  ")
    print(board[1]+" | "+ board[2] + " | "+ board[3])
    print("  |   |  ")
```

```python
def player_input():
    return input("Please pick a marker 'X' or 'O'")
```

```python
def place_marker(board, marker, position):
    board[position] = marker
    pass
```

```python
def win_check(board, mark):
    win_position = [mark, mark, mark]
    return board[1:4] == win_position or board[4: 7] == win_position or board[7:10] == win_position or board[1:8:3] == win_position or board[2:9:3] == win_position or board[3:10:3] == win_position or board[1:10:4] == win_position or board[3:8:2] == win_position
```

```python
import random

def choose_first():
    if random.randint(0, 1) == 0:
        return 'Player2'
    else:
        return 'Player1'
```

```python
def space_check(board, position):
    return board[position] == ' '
```

```python
def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
            return False
    return True 
```

```python
def player_choice(board):
    position = 0
    
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board, position):
        position = int(input('Choose your next position: (1-9) '))
        
    return position
```

```python
def replay():
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
```

```python
print('Welcome to Tic Tac Toe!')

while True: 
    marker1 = player_input()
    marker2 = 'XO'.replace(marker1, '')

    board = [' '] * 10
    display_board(board)
    turn = choose_first()

    while True: 
        print(turn)
        position =  player_choice(board)
        if turn == 'Player1':
            place_marker(board, marker1, position)
            display_board(board)
            if win_check(board, marker1): 
                print('Congratulations! You have won the game!')
                break
            else:
                if full_board_check(board):
                    display_board(board)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player2'
        else:
            place_marker(board, marker2, position)
            display_board(board)
            if win_check(board, marker2): 
                print('Congratulations! You have won the game!')
                break
            else:
                if full_board_check(board):
                    display_board(board)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player1'
    if not replay():
        break
```
